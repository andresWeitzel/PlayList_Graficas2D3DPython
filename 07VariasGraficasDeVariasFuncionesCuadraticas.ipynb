```python
import matplotlib.pyplot as plt
```

```python
x_lista01=range(0,27)
x_lista02=range(0,20)
x_lista03=range(0,10)


y_lista01=[]

y_lista02=[]

y_lista03=[]


for x in x_lista01:
    
    y= 2*(x**2) + 4*x + 6

    y_lista01.append(y)
    
    
for x in x_lista02:
    
    y= 8*(x**2) + 9*x + 9

    y_lista02.append(y)    
    
    
for x in x_lista03:
    
    y= 5*(x**2) + 9*x + 10

    y_lista03.append(y)
    
```

```python
#-------------------------GRAFICA01--------------------------------
#subplot(N°FILAS,N°COLUMNAS,N°GRAFICA)
plt.subplot(1,3,1)

#vamos a darle espaciado entre cada grafica
plt.subplots_adjust(wspace=1)

plt.plot(x_lista01, y_lista01, '--', linewidth=2)

plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=20)

leyenda=['Notebooks']

plt.legend(loc="upper left", labels=leyenda, fontsize=10)

plt.ylabel('Cantidad [Unidades]', fontsize=15)

plt.xlabel('Marzo [Días]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')


#-------------------------GRAFICA02--------------------------------
#subplot(N°FILAS,N°COLUMNAS,N°GRAFICA)
plt.subplot(1,3,2)

#vamos a darle espaciado entre cada grafica
plt.subplots_adjust(wspace=2)

plt.plot(x_lista02, y_lista02, 'r--', linewidth=2)

plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=20)

leyenda=['IPhones']

plt.legend(loc="upper left", labels=leyenda, fontsize=10)

plt.ylabel('Cantidad [Unidades]', fontsize=15)

plt.xlabel('Abril [Días]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')


#-------------------------GRAFICA03--------------------------------
#subplot(N°FILAS,N°COLUMNAS,N°GRAFICA)
plt.subplot(1,3,3)

#vamos a darle espaciado entre cada grafica
plt.subplots_adjust(wspace=2)

plt.plot(x_lista03, y_lista03, 'g--', linewidth=2)

plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=20)

leyenda=['Tv']

plt.legend(loc="upper left", labels=leyenda, fontsize=10)

plt.ylabel('Cantidad [Unidades]', fontsize=15)

plt.xlabel('Mayo [Días]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')







plt.show()
```
