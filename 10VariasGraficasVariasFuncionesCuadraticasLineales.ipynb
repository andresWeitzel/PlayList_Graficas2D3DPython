```python
import matplotlib.pyplot as plt
```

```python
x_listaFuncion01Cuadratica=range(0,27)
x_listaFuncion02Cuadratica=range(0,27)

x_listaFuncion01Lineal=range(0,27)
x_listaFuncion02Lineal=range(0,27)


y_listaFuncion01Cuadratica=[]
y_listaFuncion02Cuadratica=[]


y_listaFuncion01Lineal=[]
y_listaFuncion02Lineal=[]

```

```python
def funcionCuadratica(valores_x, valores_y, coeficientePrincipal, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:
        
        y=coeficientePrincipal * (x**2) + coeficienteLineal * x + terminoIndependiente
        
        valores_y.append(y)
        
        
def funcionLineal(valores_x, valores_y, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:
        
        y =  coeficienteLineal * x + terminoIndependiente
        
        valores_y.append(y)        
```

```python
def propiedadesGraficaDosFunciones(
    
    nroFilas, nroColumna, nroGrafica,
    valores_x_Funcion01, valores_y_Funcion01, tipoLinea_Funcion01,
    valores_x_Funcion02, valores_y_Funcion02, tipoLinea_Funcion02,anchoLinea
):
    
    plt.subplot(nroFilas,nroColumna,nroGrafica)

    plt.plot(
        valores_x_Funcion01, valores_y_Funcion01, tipoLinea_Funcion01, 
        valores_x_Funcion02, valores_y_Funcion02, tipoLinea_Funcion02, 
        linewidth=anchoLinea
    )
    
    plt.subplots_adjust(wspace=1)
    


    
    
def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):
    
    
    plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=20)

    leyenda=['Notebooks','Smartphones','Screens']

    plt.legend(loc="upper left", labels=leyenda, fontsize=10)

    plt.ylabel('Cantidad [Unidades]', fontsize=15)

    plt.xlabel(nombreEtiqueta_x,fontsize=tamañoLetras)

    plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')

    
```

```python
#-------------------------GRAFICA01--------------------------------

funcionCuadratica(x_listaFuncion01Cuadratica, y_listaFuncion01Cuadratica, 2, 4 ,6 )

funcionLineal(x_listaFuncion01Lineal, y_listaFuncion01Lineal, 9 ,9 )

#Seteamos un tamaño mas grande por grafica, con una sola declaracion basta para la salida
plt.rcParams["figure.figsize"] = (15, 5)

propiedadesGraficaDosFunciones(
    1,2,1, x_listaFuncion01Cuadratica, y_listaFuncion01Cuadratica, 'b--',
     x_listaFuncion01Lineal, y_listaFuncion01Lineal,'r--', 1 )   

#def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):
referenciasGrafica('Marzo [Días]', 15)



#-------------------------GRAFICA02--------------------------------

funcionCuadratica(x_listaFuncion02Cuadratica, y_listaFuncion02Cuadratica, 2, 4 ,6 )

funcionLineal(x_listaFuncion02Lineal, y_listaFuncion02Lineal, 9 ,9 )

propiedadesGraficaDosFunciones(
    1,2,2, x_listaFuncion02Cuadratica, y_listaFuncion02Cuadratica, 'b--',
     x_listaFuncion02Lineal, y_listaFuncion02Lineal,'r--', 1 )   

#def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):
referenciasGrafica('Marzo [Días]', 15)
```
