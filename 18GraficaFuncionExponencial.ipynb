```python
'''
IMPORTANTE: RECORDAR QUE LAS FUNCIONES CUADRATICAS Y LAS EXPONENCIALES SON COMPLETAMENTE DIFERENCTES
TANTO GRAFICA COMO ANALITICAMENTE HABLANDO. LAS CUADRATICAS IMPLEMENTAN POLINOMIOS, POR LO TANTO
SUS COEFICIENTES SON RACIONALES(-5/3 0.77), NATURALES(1,2,3) Y ENTEROS(-98-177,0,1) PERO NO 
IRRACIONALES(RAICES), EN CAMBIO LAS EXPONENCIALES INCLUYEN 'TODAS LAS CATEGORIAS', LOS REALES(RACIONALES,IRRACIONALES,NATURALES
Y ENTEROS). OJO CON ESO!!
'''

import matplotlib.pyplot as plt
import math
```

```python
x_lista=range(0,27)

y_lista=[]


for x in x_lista:
    
    '''
    -pow(x, y[, z])
    
    -Con pow elevamos el numeros deseado pasado dentro del primer parametro, y en el segundo
    parametro le indicamos a que potencia se va a elevar.
    
    -Nota: El método pow () llamada directamente a través del método integrado,  
    math.e, que sería como un entero, y los parámetros del módulo de matemáticas será 
    convertido a float. y= pow(math.e, x)
    
    -math.e invocamos a la funcion exponencial del modulo math
    '''
    
    y= pow(math.e, x)

    
    y_lista.append(y)
```

```python
plt.plot(x_lista, y_lista, '--', linewidth=2)

plt.suptitle('Ganancias 2021', fontsize=20)

leyenda=['% de Ganancia']

plt.legend(loc="upper left", labels=leyenda, fontsize=10)

plt.ylabel('Ganancia [%]', fontsize=15)

plt.xlabel('Marzo [Días]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')

plt.scatter(22, 0.05*(1E11), s=30, color='g')
plt.text(22 + 1, 0.05*(1E11), 'Primer Crecimiento')

plt.scatter(23.7, 0.25*(1E11), s=30, color='g')
plt.text(23.7 + 1, 0.25*(1E11), 'Segundo Crecimiento')

plt.scatter(25, 0.75*(1E11), s=30, color='g')
plt.text(25 + 1, 0.75*(1E11), 'Crecimiento Estabilizado')

plt.scatter(26, 2.00*(1E11), s=40, color='r')
plt.text(26 + 1, 2.00*(1E11), 'Máximo Crecimiento')
plt.text(26 + 1, 1.90*(1E11), '(Día 26)')


#Seteamos un tamaño mas grande por grafica, con una sola declaracion basta para la salida
plt.rcParams["figure.figsize"] = (18, 10)

```
