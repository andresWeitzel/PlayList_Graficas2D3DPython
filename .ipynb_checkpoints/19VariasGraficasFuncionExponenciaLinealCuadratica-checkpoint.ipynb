```python
import matplotlib.pyplot as plt
import math
```

```python
x_listaFuncionCuadratica=range(0,27)
x_listaFuncionLineal=range(0,27)
x_listaFuncionExponencial=range(0,27)

y_listaFuncionCuadratica=[]
y_listaFuncionLineal=[]
y_listaFuncionExponencial=[]

```

```python


def funcionCuadratica(valores_x, valores_y, coeficientePrincipal, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:

        y = coeficientePrincipal*(x**2) + coeficienteLineal*x + terminoIndependiente

        valores_y.append(y)
        

        
def funcionLineal(valores_x, valores_y, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:
        
        y = coeficienteLineal*x + terminoIndependiente
        
        valores_y.append(y)

        
def funcionExponencial(valores_x, valores_y):       

    for x in valores_x:

        y= pow(math.e, x)

        valores_y.append(y)        
```

```python
def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
    
    plt.subplot(nroFilas,nroColumna,nroGrafica)

    plt.subplots_adjust(wspace=0.98,hspace=0.5)
    
    plt.plot(valores_x, valores_y, tipoLinea, linewidth=anchoLinea)

    

def referenciasGrafica(
    nombreEtiqueta_x, tamañoNombreEtiqueta_x, 
    nombreEtiqueta_y, tamañoNombreEtiqueta_y,
    leyenda, tamañoLeyenda
):
 
    plt.suptitle('Ganancias 2021', fontsize=20)

    leyendaLista=[leyenda]

    plt.legend(loc="upper left", labels=leyendaLista, fontsize=tamañoLeyenda)

    plt.ylabel(nombreEtiqueta_y, fontsize=tamañoNombreEtiqueta_y)

    plt.xlabel(nombreEtiqueta_x,fontsize=tamañoNombreEtiqueta_x)

    plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')
    
    
    
def referenciasAnalitica(posicion_x, posicion_y, tamaño, color, etiquetaTexto ):  

    plt.scatter(posicion_x, posicion_y, s=tamaño, color=color)
    
    plt.text(posicion_x + 2, posicion_y, etiquetaTexto)
    

    
```

```python
funcionCuadratica(x_listaFuncionCuadratica, y_listaFuncionCuadratica, 2, 4 ,6 )
funcionLineal(x_listaFuncionLineal, y_listaFuncionLineal, 8, 9 )
funcionExponencial(x_listaFuncionExponencial, y_listaFuncionExponencial)
```

```python


#-------------------------GRAFICA01--------------------------------

#def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
propiedadesGrafica(1,3,1, x_listaFuncionLineal, y_listaFuncionLineal, '-', 2 )   

#def referenciasGrafica(nombreEtiqueta_x, tamañoNombreEtiqueta_x,nombreEtiqueta_y, tamañoNombreEtiqueta_y, leyenda, tamañoLeyenda):
referenciasGrafica('Enero [Días]', 10, 'Ganancia Bruto [%]', 10, '% Ganancia Bruto', 8 )

 #Le damos un limite con una linea en vertical
plt.axvline(26.3, color="red",linewidth=1,linestyle="dashed")
  

#Seteamos un tamaño mas grande por grafica, con una sola declaracion basta para la salida
plt.rcParams["figure.figsize"] = (17, 5)



#-------------------------GRAFICA02--------------------------------

#def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
propiedadesGrafica(1,3,2, x_listaFuncionCuadratica, y_listaFuncionCuadratica, '-', 2 )   

#def referenciasGrafica(nombreEtiqueta_x, tamañoNombreEtiqueta_x,nombreEtiqueta_y, tamañoNombreEtiqueta_y, leyenda, tamañoLeyenda):
#La ganancia Operativa es la ganancia sin los impuestos, etc.
referenciasGrafica('Febrero [Días]', 10, 'Ganancia Operativa [%]', 10, '% Ganancia Operativa', 8 )

 #Le damos un limite con una linea en vertical
plt.axvline(26.3, color="red",linewidth=1,linestyle="dashed")
  

#-------------------------GRAFICA03--------------------------------

#def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
propiedadesGrafica(1,3,3, x_listaFuncionExponencial, y_listaFuncionExponencial, '-', 2 )   

#def referenciasGrafica(nombreEtiqueta_x, tamañoNombreEtiqueta_x,nombreEtiqueta_y, tamañoNombreEtiqueta_y, leyenda, tamañoLeyenda):
referenciasGrafica('Marzo [Días]', 10, 'Ganancia Neta [%]', 10, '% Ganancia Neta', 8)

#def referenciasAnalitica(posicion_x, posicion_y, tamaño, color, etiquetaTexto ): 
referenciasAnalitica(21, 0.03*(1E11), 30, 'g', 'Primer Crecimiento')
referenciasAnalitica(25, 0.75*(1E11), 30, 'g', 'Crecimiento Estabilizado')
referenciasAnalitica(26, 2.00*(1E11), 60, 'r', 'Máximo Crecimiento\n            (Día 26)')

#Le damos un limite con una linea en vertical
plt.axvline(26.3, color="red",linewidth=1,linestyle="dashed")

#Le damos un limite con una linea en horizontal
#Le damos un techo y piso para un rango determinado, como por ejemplo el % minimo y maximo esperado
plt.axhline(0.03*(1E11), color="green",linewidth=0.7,linestyle="dashed")
plt.axhline(0.75*(1E11), color="green",linewidth=0.7,linestyle="dashed") 

'''
Podemos concluir que el porcentaje de ganancia neta es mas bajo que los otros, ya que es
el porcentaje real de ganancia, y traspolandolo a la grafica la curva de crecimiento es casi
igual hasta el dia 20
'''




   
```
