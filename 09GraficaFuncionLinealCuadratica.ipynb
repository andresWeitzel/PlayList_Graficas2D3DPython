```python
import matplotlib.pyplot as plt
```

```python
x_listaFuncionCuadratica=range(0,20)
x_listaFuncionLineal=range(0,20)

y_listaFuncionCuadratica=[]
y_listaFuncionLineal=[]
```

```python

def funcionCuadratica(valores_x, valores_y, coeficientePrincipal, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:

        y = coeficientePrincipal*(x**2) + coeficienteLineal*x + terminoIndependiente

        valores_y.append(y)
        

        
def funcionLineal(valores_x, valores_y, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:
        
        y = coeficienteLineal*x + terminoIndependiente
        
        valores_y.append(y)
```

```python
        
funcionCuadratica(x_listaFuncionCuadratica, y_listaFuncionCuadratica, 3, 6, 4)


funcionLineal(x_listaFuncionLineal, y_listaFuncionLineal, 3, 2)
```

```python
def propiedadesGrafica(nroFilas, nroColumnas, nroFigura, valores_x, valores_y, tipoLinea, anchoLinea):
    
    plt.subplot(nroFilas,nroColumnas,nroFigura)
    
    plt.plot(valores_x, valores_y,tipoLinea, linewidth=anchoLinea)
    

    
def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):

    plt.subplots_adjust(wspace=1)
    
    plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=15)

    leyendas=['Notebooks','Smartphones','Screens']

    plt.legend(loc="upper left", labels=leyendas, fontsize=10)

    plt.ylabel('Cantidad [Unidades]',fontsize=15)

    plt.xlabel(nombreEtiqueta_x,fontsize=tamañoLetras)

    plt.grid(True, linestyle='-.', linewidth=0.5 ,color='gray')
    


    
```

```python
#--------------------FIGURA 01-----------------------------------

#def propiedadesGrafica(nroFilas, nroColumnas, nroFigura, valores_x, valores_y, tipoLinea, anchoLinea):
propiedadesGrafica(1, 2, 1, x_listaFuncionCuadratica, y_listaFuncionCuadratica, 'b-' , 1 )

#def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):
referenciasGrafica('Enero [Días]',15)

#--------------------FIGURA 02-----------------------------------

propiedadesGrafica(1, 2, 2, x_listaFuncionLineal, y_listaFuncionLineal, 'b--' , 3 )

referenciasGrafica('Febrero [Días]',15)

```
