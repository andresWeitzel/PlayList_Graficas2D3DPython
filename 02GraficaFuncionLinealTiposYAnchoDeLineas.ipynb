```python
import matplotlib.pyplot as plt
```

```python
x_lista=[1,2,3,4]

x_lista=list(map(int,x_lista))

y_lista=[]

###########################

for i in x_lista:
    
    y= 3*i + 2

    y_lista.append(y)


##########################

plt.plot(x_lista, y_lista,'o')
```
