```python
import matplotlib.pyplot as plt
```

```python
x_lista=[1,2,3,4]

x_lista=list(map(int,x_lista))

y_lista=[]

###########################

for i in x_lista:
    
    y= 3*i + 2

    y_lista.append(y)


##########################

plt.plot(x_lista, y_lista, '--', linewidth=1)

plt.ylabel('Eje de las Ordenadas')

plt.xlabel('Eje de las Abscisas')
```

```python
plt.plot(x_lista, y_lista, '--', linewidth=1)

plt.ylabel('Eje de las Ordenadas', fontsize=15)

plt.xlabel('Eje de las Abscisas',fontsize=15)
```

```python
plt.plot(x_lista, y_lista, '--', linewidth=1)

plt.suptitle('Grfica Funcin Lineal', fontsize=20)

leyenda=['Grfica01']

plt.legend(loc="upper left", labels=leyenda, fontsize=15)

plt.ylabel('Eje de las Ordenadas', fontsize=15)

plt.xlabel('Eje de las Abscisas',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')

#El .show() al final para que no se nos genere algn error
#con las propiedades agregadas

plt.show()
```

```python
plt.plot(x_lista, y_lista, '--', linewidth=1)

plt.suptitle('Movimiento Rectilneo Uniforme', fontsize=20)

leyenda=['Grfica01']

plt.legend(loc="upper left", labels=leyenda, fontsize=15)

plt.ylabel('Velocidad[m/s]', fontsize=15)

plt.xlabel('Tiempo[s]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')

#El .show() al final para que no se nos genere algn error
#con las propiedades agregadas

plt.show()
```
