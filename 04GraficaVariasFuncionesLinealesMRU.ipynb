```python
import matplotlib.pyplot as plt
```

```python
x_lista=[1,2,3,4]
x_lista02=[3.5,4.5,5.5,6.5]

y_lista=[]
y_lista02=[]

###########################

for i in x_lista:
    
    y= 3*i + 2

    y_lista.append(y)

    
for i in x_lista02:
    
    y= 6*i + 16

    y_lista02.append(y)
    
    


##########################

```

```python
plt.plot(x_lista, y_lista, '--', x_lista02, y_lista02, 'r--', linewidth=1)

plt.suptitle('Movimiento Rectilíneo Uniforme', fontsize=20)

leyenda=['Gráfica01','Gráfica02']

plt.legend(loc="upper left", labels=leyenda, fontsize=15)

plt.ylabel('Velocidad[m/s]', fontsize=15)

plt.xlabel('Tiempo[s]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')

#El .show() al final para que no se nos genere algún error
#con las propiedades agregadas

plt.show()
```
