```python
import numpy as np
import matplotlib.pyplot as plt
    
'''
Para generar funciones aleatorias que usaremos luego para producir componenetes de 
colores RGB
'''
import random
```

```python
#plt.axis([x1,x2,y1,y2])
plt.axis([0,200,0,100])
plt.grid(True)
plt.title('Gráfica de Puntos')

#plt.arrow('x','y','dx','dy')->dx/dy son las longitudes de las flechas por coordenadas
#head_length=4 y head_width es para la punta del vector
plt.arrow(10,90,20,0, head_length=4, head_width=3)#Abscisas 
plt.arrow(10,90,0,-15, head_length=4, head_width=3)#Ordenadas

plt.text(15,95,'x')
plt.text(3,80,'y')


'''
CONSTRUIMOS PLOTEOS DE PUNTOS EN DIVERSAS GRAFICAS 
'''

#====================GRÁFICA 01=====================

#np.arange('valor inicial','valor final','incremento')
for x in np.arange(50,75,6):
    for y in np.arange(20,80,6):
        
        #plt.scatter(x(array), y(array), s(tamaño) color, etc))
        plt.scatter(x, y, s=2, color='b')#k is black / b is blue
    
    
#====================GRÁFICA 02=====================    
    
for x in np.arange(100,125,.5):
    for y in np.arange(20,80,.5):
        
        #plt.scatter(x(array), y(array), s(tamaño) color, etc))
        plt.scatter(x, y, s=2, color='b')#k is black / b is blue
        
        
        
'''
        Recordar que los canales RBG
        van como bit maximo 256,
        pero aca se usa cmap,
        segun matplotlib...Si c es una matriz de flotantes,
        se usa norm para escalar
        los datos de color, c, en el rango de 0 a 1, con el fin de 
        mapear en el mapa 
        de colores cmap. Si es Ninguno, use los colores predeterminados. 
        Normalizar.
        Normalizamos trabajando de 0 a 100 y dividiendo el posible valor
        aleatorio sobre 100
        (NINGUN VALOR VA A SOBREPASAR EL 1)
''' 


#====================GRÁFICA 03=====================    
    
for x in np.arange(150,175,2.5):
    for y in np.arange(20,80,2.5):
        
        #randrange(start, stop, step)
        R=random.randrange(0,100,1)/100
        G=random.randrange(0,100,1)/100
        B=random.randrange(0,100,1)/100
        
        #plt.scatter(x(array), y(array), s(tamaño) color, etc))
        plt.scatter(x, y, s=20, color=(R,G,B))#k is black / b is blue
        
     
```
