```python
'''
==================================================================
=====GRAFICA DE PUNTOS, LINEAS RECTAS  | PRIMERA FORMA DE IMPLEMENTACION=======
==============RAZONAMIENTO MATEMATICO=============================
==================================================================

'''
#plt.axis([x1,x2,y1,y2])
plt.axis([-10,140,100,-10])
#plt.arrow('x','y','dx','dy')
plt.arrow(0,0,20,0,head_length=4,head_width=3,color='k')#abcisas
plt.arrow(0,0,0,20,head_length=4,head_width=3,color='k')#Ordenadas

plt.text(15, -3, 'x')
plt.text(-5, 15, 'y')


plt.axis('on')
plt.grid(True)
plt.title('LINEA RECTAS')


#===========Generaremos nuestra Linea de Puntos por coordenadas==============

#Coordenadas de los puntos(Punto inicial 1, Punto final 2)
x1=20
x2=120
y1=40
y2=20



#Longitud total de la linea
#Q = sqrt[(x2 - x1)*2 + (y2 - y1)*2 ]
longitud=np.sqrt((x2-x1)**2+(y2-y1)**2)

#Escalares de los ejes
ux=(x2-x1) / longitud
uy=(y2-y1) / longitud 

#Marcamos los puntos en la grafica
for l in np.arange(0, longitud , 1):
    px=x1+l*ux
    py=y1+l*uy
     #plt.scatter(x_array, y_array, s_tamao, color, etc)
    plt.scatter(px,py,s=1,color='g')

    
#Vamos a marcar los puntos en la grafica, el uno y el dos
plt.scatter(x1, y1, s=20, color='k')
plt.text(x1 + 3, y1 + 3, 'Punto Inicial')

plt.scatter(x2, y2, s=20, color='k')
plt.text(x2 + 3, y2 + 3, 'Punto Final')

#Seteamos un tamao mas grande por grafica, con una sola declaracion basta para la salida
plt.rcParams["figure.figsize"] = (17, 10)

```
