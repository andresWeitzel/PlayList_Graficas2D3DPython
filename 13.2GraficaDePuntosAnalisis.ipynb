```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def limitesEjes():
    #plt.axis([x1,x2,y1,y2])
    plt.axis([0,200,0,200])
    plt.grid(True)
    plt.title('Gráfica de Puntos')
    
```

```python
limitesEjes()

#================ Diseñar una linea ==========
for x in np.arange(50,175,0.1):
    for y in np.arange(20,21,6):
        
        #plt.scatter(x(array), y(array), s(tamaño) color, etc))
        plt.scatter(x, y, s=2, color='b')#k is black / b is blue
    
```

```python
def generateLines(x1,x2,y1,y2):
    for x in np.arange(x1,x2,1):
        for y in np.arange(y1,y2,1):
            plt.scatter(x, y, s=2, color='b')#k is black / b is blue
    
```

```python
limitesEjes()
generateLines(50,175,20,21)
generateLines(50,175,75,76)

generateLines(50,51,20,76)
generateLines(175,176,20,76)
```
