```python
import matplotlib.pyplot as plt
```

```python
x_lista01=range(0,27)
x_lista02=range(0,20)
x_lista03=range(0,10)


y_lista01=[]

y_lista02=[]

y_lista03=[]


for x in x_lista01:
    
    y= 2*(x**2) + 4*x + 6

    y_lista01.append(y)
    
    
for x in x_lista02:
    
    y= 8*(x**2) + 9*x + 9

    y_lista02.append(y)    
    
    
for x in x_lista03:
    
    y= 5*(x**2) + 9*x + 10

    y_lista03.append(y)
```

```python
plt.plot(x_lista01, y_lista01, '--', x_lista02, y_lista02, 'r--', x_lista03, y_lista03, 'g--', linewidth=2)

plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=20)

leyenda=['Notebooks','Celulares','Accesorios']

plt.legend(loc="upper left", labels=leyenda, fontsize=10)

plt.ylabel('Cantidad [Unidades]', fontsize=15)

plt.xlabel('Marzo [Días]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')

#El .show() al final para que no se nos genere algún error
#con las propiedades agregadas

plt.show()
```
