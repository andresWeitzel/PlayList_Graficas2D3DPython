```python
import matplotlib.pyplot as plt
```

```python
x_listaFuncion01Cuadratica=range(0,27)
x_listaFuncion01Lineal=range(0,27)

x_listaFuncion02Cuadratica=range(0,27)
x_listaFuncion02Lineal=range(0,27)


y_listaFuncion01Cuadratica=[]
y_listaFuncion01Lineal=[]

y_listaFuncion02Cuadratica=[]
y_listaFuncion02Lineal=[]
```

```python

def funcionCuadratica(valores_x, valores_y, coeficientePrincipal, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:

        y = coeficientePrincipal*(x**2) + coeficienteLineal*x + terminoIndependiente

        valores_y.append(y)
        

        
def funcionLineal(valores_x, valores_y, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:
        
        y = coeficienteLineal*x + terminoIndependiente
        
        valores_y.append(y)

        
```

```python

def propiedadesGrafica(nroFilas, nroColumnas, nroFigura, valores_x, valores_y, tipoLinea, anchoLinea):
    
    plt.subplot(nroFilas,nroColumnas,nroFigura)
    
    plt.plot(valores_x, valores_y,tipoLinea, linewidth=anchoLinea)
    

    
def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):

    plt.subplots_adjust(wspace=1)
    
    plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=15)

    leyendas=['Notebooks','Smartphones','Screens']

    plt.legend(loc="upper left", labels=leyendas, fontsize=10)

    plt.ylabel('Cantidad [Unidades]',fontsize=15)

    plt.xlabel(nombreEtiqueta_x,fontsize=tamañoLetras)

    plt.grid(True, linestyle='-.', linewidth=0.5 ,color='gray')
    

'''
Logicamente que al crear funciones estas se quedan limitadas en cuanto a su funcionamiento.
Por eso es que cuanto mas cantidad de funciones grafiquemos, aumentara la cantidad de parametros
a detallar en el argumento de nuestra funcion creada, vamos a sobrecargar nuestras funciones

'''    

def propiedadesGraficaDosFunciones(
    
    nroFilas, nroColumnas, nroFigura,
    valores_x_Funcion01, valores_y_Funcion01, tipoLinea_Funcion01,
    valores_x_Funcion02, valores_y_Funcion02, tipoLinea_Funcion02, anchoLinea_Funcion02
):
    
    
    plt.subplot(nroFilas,nroColumnas,nroFigura)
    
    plt.plot(
        valores_x_Funcion01, valores_y_Funcion01, tipoLinea_Funcion01, 
        valores_x_Funcion02, valores_y_Funcion02, tipoLinea_Funcion02, linewidth=anchoLinea_Funcion02
    )
    
    
    
    #Le damos un limite con una linea en vertical
    plt.axvline(27, color="black",linewidth=1,linestyle="dashed")
    
 

    
def referenciasGraficaDosFunciones(nombreEtiqueta_x, tamañoLetras):

   
    
    plt.subplots_adjust(wspace=1)
    
    plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=15)

    leyendas=['Notebooks','Smartphones','Screens']

    plt.legend(loc="upper left", labels=leyendas, fontsize=10)

    plt.ylabel('Cantidad [Unidades]',fontsize=15)

    plt.xlabel(nombreEtiqueta_x,fontsize=tamañoLetras)

    plt.grid(True, linestyle='-.', linewidth=0.5 ,color='gray')
    
    plt.style.available
    
    plt.style.use("seaborn-deep")
    
    '''
     'Solarize_Light2',
     '_classic_test_patch',
     'bmh',
     'classic',
     'dark_background',
     'fast',
     'fivethirtyeight',
     'ggplot',
     'grayscale',
     'seaborn',
     'seaborn-bright',
     'seaborn-colorblind',
     'seaborn-dark',
     'seaborn-dark-palette',
     'seaborn-darkgrid',
     'seaborn-deep',
     'seaborn-muted',
     'seaborn-notebook',
     'seaborn-paper',
     'seaborn-pastel',
     'seaborn-poster',
     'seaborn-talk',
     'seaborn-ticks',
     'seaborn-white',
     'seaborn-whitegrid',
     'tableau-colorblind10'
    '''
    
    
```
