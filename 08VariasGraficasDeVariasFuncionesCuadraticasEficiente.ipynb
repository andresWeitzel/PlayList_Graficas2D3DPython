```python
import matplotlib.pyplot as plt
```

```python
x_lista01=range(0,27)
x_lista02=range(0,20)
x_lista03=range(0,10)


y_lista01=[]

y_lista02=[]

y_lista03=[]
```

```python

def funcionCuadratica(valores_x, valores_y, coeficientePrincipal, coeficienteLineal, terminoIndependiente):
    
    for x in valores_x:
        
        y=coeficientePrincipal * (x**2) + coeficienteLineal * x + terminoIndependiente
        
        valores_y.append(y)
```

```python
funcionCuadratica(x_lista01,y_lista01, 2, 4 ,6 )
funcionCuadratica(x_lista02,y_lista02, 8, 9 ,9 )
funcionCuadratica(x_lista03,y_lista03, 5, 9 ,10 )
```

```python
def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
    
    plt.subplot(nroFilas,nroColumna,nroGrafica)

    plt.plot(valores_x, valores_y, tipoLinea, linewidth=anchoLinea)
    
    plt.subplots_adjust(wspace=1)
    

def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):
    
    
    plt.suptitle('Venta de Elementos Electrónicos 2021', fontsize=20)

    leyenda=['Notebooks']

    plt.legend(loc="upper left", labels=leyenda, fontsize=10)

    plt.ylabel('Cantidad [Unidades]', fontsize=15)

    plt.xlabel(nombreEtiqueta_x,fontsize=tamañoLetras)

    plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')

    
```

```python
#-------------------------GRAFICA01--------------------------------
#def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
propiedadesGrafica(1,3,1, x_lista01, y_lista01, '--', 1 )   

#def referenciasGrafica(nombreEtiqueta_x, tamañoLetras):
referenciasGrafica('Marzo [Días]', 15)

#-------------------------GRAFICA02--------------------------------

propiedadesGrafica(1,3,2, x_lista02, y_lista02, 'r--', 1 )   

referenciasGrafica('Abril [Días]', 15)

#-------------------------GRAFICA03--------------------------------
propiedadesGrafica(1,3,3, x_lista03, y_lista03, 'g--', 1 )   

referenciasGrafica('Mayo [Días]', 15)



plt.show()

```
