```python
import matplotlib.pyplot as plt
```

```python
#Establecemos rango de 10 d√≠as
x_lista01=range(0,10)
x_lista02=range(0,10)
#Variamos a un rango de 20 dias
x_lista03=range(0,20)

y_lista01=[]
y_lista02=[]
y_lista03=[]

###########################

for x in x_lista01:
    
    y= 3*(x**2) + 2*x + 3

    y_lista01.append(y)


for x in x_lista02:
    
    y= 5*(x**2) + 8*x + 9

    y_lista02.append(y)


for x in x_lista03:
    
    y= 20*(x**2) + 10*x + 20

    y_lista03.append(y)
    
##########################


```

```python
#subplot(N¬∞FILAS,N¬∞COLUMNAS,N¬∞FIGURA)
plt.subplot(1,3,1)

#Vamos a darle espaciado entre cada figura
plt.subplots_adjust(wspace=1)

plt.plot(x_lista01, y_lista01, 'b-', linewidth=1)

plt.suptitle('Venta de Elementos Electr√≥nicos 2021', fontsize=15)

leyendas=['Notebooks','Smartphones','Screens']

plt.legend(loc="upper left", labels=leyendas, fontsize=10)

plt.ylabel('Cantidad [Unidades]',fontsize=15)

plt.xlabel('Marzo [D√≠as]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5 ,color='gray')


#####################################################################

plt.subplot(1,3,2)

#Vamos a darle espaciado entre cada figura
plt.subplots_adjust(wspace=1)

plt.plot(x_lista02, y_lista02, 'r--', linewidth=2)

leyendas=['Notebooks','Smartphones','Screens']

plt.legend(loc="upper left", labels=leyendas, fontsize=10)

plt.ylabel('Cantidad [Unidades]',fontsize=15)

plt.xlabel('Abril [D√≠as]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5 ,color='gray')

####################################################################

plt.subplot(1,3,3)

plt.plot(x_lista03, y_lista03, 'g--', linewidth=3)

leyendas=['Notebooks','Smartphones','Screens']

plt.legend(loc="upper left", labels=leyendas, fontsize=10)

plt.ylabel('Cantidad [Unidades]',fontsize=15)

plt.xlabel('Mayo [D√≠as]',fontsize=15)

plt.grid(True, linestyle='-.', linewidth=0.5 ,color='gray')

####################################################################


plt.show()
```
