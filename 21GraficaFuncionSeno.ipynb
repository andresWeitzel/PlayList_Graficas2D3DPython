```python
import matplotlib.pyplot as plt
#numpy para implementar una matriz con incrementos para las funciones
import numpy as np

import math
```

```python
#Creamos un array en vez de un rango, para trabajar con floats y enteros
x_listaFuncionSeno=np.arange(0,20,0.1)

y_listaFuncionSeno=[]
```

```python
def funcionSeno(valores_x, valores_y):
     
        for x in valores_x:

            y = math.sin(x)

            valores_y.append(y)
```

```python
def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
    
    plt.subplot(nroFilas,nroColumna,nroGrafica)

    plt.subplots_adjust(wspace=0.98,hspace=0.5)
    
    plt.plot(valores_x, valores_y, tipoLinea, linewidth=anchoLinea)

    

def referenciasGrafica(
    nombreEtiqueta_x, tamañoNombreEtiqueta_x, 
    nombreEtiqueta_y, tamañoNombreEtiqueta_y,
    leyenda, tamañoLeyenda
):
 
    plt.suptitle('Medición Eléctrica Baja Tensión', fontsize=20)

    leyendaLista=[leyenda]

    plt.legend(loc="upper left", labels=leyendaLista, fontsize=tamañoLeyenda)

    plt.ylabel(nombreEtiqueta_y, fontsize=tamañoNombreEtiqueta_y)

    plt.xlabel(nombreEtiqueta_x,fontsize=tamañoNombreEtiqueta_x)

    plt.grid(True, linestyle='-.', linewidth=0.5, color='gray')
    
    
    
def referenciasAnalitica(posicion_x, posicion_y, tamaño, color, etiquetaTexto ):  

    plt.scatter(posicion_x, posicion_y, s=tamaño, color=color)
    
    plt.text(posicion_x + 1, posicion_y, etiquetaTexto)
    
```

```python
funcionSeno(x_listaFuncionSeno, y_listaFuncionSeno)
```

```python


#-------------------------GRAFICA01--------------------------------

#def propiedadesGrafica(nroFilas, nroColumna, nroGrafica, valores_x, valores_y, tipoLinea, anchoLinea):
propiedadesGrafica(1,1,1, x_listaFuncionSeno, y_listaFuncionSeno, '-', 2 )   


#Seteamos un tamaño mas grande por grafica, con una sola declaracion basta para la salida
plt.rcParams["figure.figsize"] = (20, 5)

#def referenciasGrafica(nombreEtiqueta_x, tamañoNombreEtiqueta_x,nombreEtiqueta_y, tamañoNombreEtiqueta_y, leyenda, tamañoLeyenda):
referenciasGrafica('Tiempo [Segundos]', 10, 'Voltaje [volts]', 10, 'Voltaje [Volts]', 8 )

#def referenciasAnalitica(posicion_x, posicion_y, tamaño, color, etiquetaTexto ): 
referenciasAnalitica(7.6, 1.00, 30, 'r', 'Voltaje Máximo')
referenciasAnalitica(11, -1.00, 30, 'r', 'Voltaje Mínimo')

#Le damos un limite con una linea en horizontal
plt.axhline(0, color="blue",linewidth=1,linestyle="dashed")
plt.axvline(20.0, color="black",linewidth=0.5,linestyle="solid")


```
