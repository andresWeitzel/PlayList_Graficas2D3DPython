```python
x=[1,2,3,4]

y_lista=[]

###########################
y= 3*x[0] + 2

y_lista.append(y)

y= 3*x[1] + 2

y_lista.append(y)

y= 3*x[2] + 2

y_lista.append(y)

y= 3*x[3] + 2

y_lista.append(y)

##########################

plt.plot(x,y_lista)
```

```python
x_lista=[1,2,3,4]

x_lista=list(map(int,x_lista))

y_lista=[]

###########################

for i in x_lista:
    
    y= 3*i + 2

    y_lista.append(y)


##########################

plt.plot(x_lista, y_lista)
```
